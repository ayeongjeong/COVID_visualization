# EDA

CARD DF
- Index(['receipt_dttm', 'adstrd_code', 'adstrd_nm', 'mrhst_induty_cl_code', 'mrhst_induty_cl_nm', 'selng_cascnt', 'salamt'],       dtype='object')
- 업종 별 결재금액 데이터 : 카드회사가 카드사용내역을 접수한일자, 가맹점 위치 기준 행정동 코드, 가맹점 위치 기준 행정동명, 가맹점 업종코드, 가맹점 업종명, 매출발생건수, 매출발생금액

매출발생 건수, 매출발생 금액을 분류로 하여 월별 결제 금액 변동 그래프 그리기 : 선제 되어야 할 것은 접수 일자를 연/월/일로 나누기.

```python
# 접수일자 -> 연/월/일

card['receipt_year'] = (card['receipt_dttm'].astype(str).str[0:4]).astype('int64')
card['receipt_month'] = (card['receipt_dttm'].astype(str).str[4:6]).astype('int64')
card['receipt_day'] = (card['receipt_dttm'].astype(str).str[6:8]).astype('int64')
card
```

## 업종별 판매 금액

```python
# 데이터 인스턴스에 담기

ind_salamt = card.groupby(['mrhst_induty_cl_nm'])['salamt'].sum().reset_index(drop = False)
ind_salamt
```

```python
ind_nm = card.mrhst_induty_cl_nm.unique()
ind_nm
```

```python
ind_salamt['sale'] = ind_salamt['salamt'].astype(float)
ind_salamt.dtypes
```

```python
ind_salamt.info()
```

```python
plt.figure()

group_nm = ind_nm
group_sz = pd.to_numeric(ind_salamt['salamt'])

plt.pie(group_sz, labels = group_nm, autopct = '%1.2f%%')

plt.show()
```
